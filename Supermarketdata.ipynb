```python
import pandas as pd
import numpy as np
```

## Create Dataset

### Create Sale Table

```python
# create date range
dates = pd.date_range(start='2023-01-01', end='2023-04-01', freq='D').tolist()

# create empty dataframe
df = pd.DataFrame(columns=['Date', 'Productdesc', 'Sales'])

# Iterate through every day
for date in dates:
    # create list of products
    products = ['Apple', 'Banana', 'Tomato', 'Orange', 'Potato']
    # Iteriere durch jeden Produkt und füge es dem DataFrame hinzu
    for product in products:
        df = df.append({'Date': date, 'Productdesc': product, 'Sales': 0}, ignore_index=True)
```

```python
#df
df_sales = df
#df_sales
```

```python
# Define Sales ranges
sales_ranges = {'Apple': (60, 100),
                'Banana': (70, 90),
                'Tomato': (40, 80),
                'Orange': (35, 60),
                'Potato': (10, 30)}

# Generate random sales numbers
for index, row in df_sales.iterrows():
    product = row['Productdesc']
    sales_range = sales_ranges[product]
    sales = np.random.randint(sales_range[0], sales_range[1]+1)
    df_sales.at[index, 'Sales'] = sales
```

### Append Inventory Column

```python
df_sales['Inventory'] = np.random.uniform(low=df_sales['Sales'], high=df_sales['Sales']*1.1).round().astype(int)
```
