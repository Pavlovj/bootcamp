## Create Dataset

### Create Sale Table

```python
# create date range
dates = pd.date_range(start='2023-01-01', end='2023-04-01', freq='D').tolist()

# create empty dataframe
df_store1 = pd.DataFrame(columns=['Date', 'Productdesc', 'Sales'])

# Iterate through every day
for date in dates:
    # create list of products
    products = ['Apple', 'Banana', 'Tomato', 'Orange', 'Potato']
    # Iteriere durch jeden Produkt und füge es dem DataFrame hinzu
    for product in products:
        df_store1 = df_store1.append({'Date': date, 'Productdesc': product, 'Sales': 0}, ignore_index=True)
```

```python
# Define Sales ranges
sales_ranges = {'Apple': (60, 100),
                'Banana': (70, 90),
                'Tomato': (40, 80),
                'Orange': (35, 60),
                'Potato': (10, 30)}

# Generate random sales numbers
for index, row in df_store1.iterrows():
    product = row['Productdesc']
    sales_range = sales_ranges[product]
    sales = np.random.randint(sales_range[0], sales_range[1]+1)
    df_store1.at[index, 'Sales'] = sales
```

### Append Inventory Column

```python
df_store1['Inventory'] = np.random.uniform(low=df_store1['Sales'], high=df_store1['Sales']*1.1).round().astype(int)
```

# Create Dataframe Store 2

```python

# create empty dataframe
df_store2 = pd.DataFrame(columns=['Date', 'Productdesc', 'Sales'])

# Iterate through every day
for date in dates:
    # create list of products
    products = ['Apple', 'Banana', 'Tomato', 'Orange', 'Potato']
    # Iteriere durch jeden Produkt und füge es dem DataFrame hinzu
    for product in products:
        df_store2 = df_store2.append({'Date': date, 'Productdesc': product, 'Sales': 0}, ignore_index=True)
```

```python
# Define Sales ranges
sales_ranges = {'Apple': (230, 260),
                'Banana': (200, 240),
                'Tomato': (260, 290),
                'Orange': (135, 160),
                'Potato': (100, 130)}

# Generate random sales numbers
for index, row in df_store2.iterrows():
    product = row['Productdesc']
    sales_range = sales_ranges[product]
    sales = np.random.randint(sales_range[0], sales_range[1]+1)
    df_store2.at[index, 'Sales'] = sales
```

```python
# add inventory store 2
df_store2['Inventory'] = np.random.uniform(low=df_store2['Sales'], high=df_store2['Sales']*1.3).round().astype(int)
```

```python
#show store 2
df_apple_2 = df_store2[df_store2['Productdesc'] == "Apple"]
#df_store2.head(20)

df_apple_2
```

# Prediction Holt Winters

```python
#load the libraries
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.filters.hp_filter import hpfilter
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from math import sqrt

plt.style.use('default')
```

# split into train and test

```python
apple_sales = df_apple_2[["Date","Sales"]]
apple_sales
```

```python
#reset the day of each timestamp in df_ez to 1
apple_sales["Date"] = apple_sales["Date"].dt.to_period('D').dt.to_timestamp()
```

```python
apple_sales.head(10)
```

```python
#sort the data by timestamp
apple_sales = apple_sales.sort_values(by=['Date'])

#reset the index
apple_sales = apple_sales.reset_index(drop=True)

apple_sales = apple_sales.set_index('Date')

apple_sales_pred = apple_sales

apple_sales_pred.info()
```

```python
#apple_sales_pred.head(10)
apple_sales_pred.plot()
```

```python
result_datadriven = seasonal_decompose(apple_sales_pred["Sales"], model='additive', period=12)
result_datadriven.plot()
```

```python
# split into train and test

train = apple_sales_pred[:len(apple_sales_pred)-12]
test = apple_sales_pred[len(apple_sales_pred)-12:]
```

```python
test
```

```python
apple_sales_pred.info()
```

```python
#change sales to int
apple_sales_pred['Sales'] = apple_sales_pred['Sales'].astype(int)
```

```python
holt_winters_model = ExponentialSmoothing(apple_sales_pred["Sales"], trend='mul', seasonal ='add', seasonal_periods = 12).fit()
```

```python
test_pred = holt_winters_model.forecast(steps = 12)
```

```python
test_pred
```

```python
train['Sales'].plot(legend=True, label = 'Train', figsize=(10,6))
test['Sales'].plot(legend= True, label = 'Test')

test_pred.plot(legend = True, label= 'predicted_test')
```
